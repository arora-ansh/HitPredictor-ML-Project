```python
import pandas as pd
import numpy as np
data = pd.read_csv('datasets/spotify_music.csv')
```

```python
data.describe()
```

```python
features = ['Explicit','Artist_followers','Track_number_on_album','Acousticness','Danceability','Energy','Liveness','Loudness','Speechiness','Tempo','Mode','Key','Valence']
label = ['Rank']
```

```python
data_features = data[features]
```

```python
data_features = (data_features - data_features.mean())/(data_features.std())
```

```python
data_features.describe()
```

```python
data_features.to_csv('datasets/spotofy_music_normalized.csv', index=False)
```

```python
def convert_score(x):
    y = []
    for i in range(len(x)):
        y.append([(199.7-0.7*x[i][0])/199])
        # y.append([201-x[i][0]])
    return np.array(y)
```

```python
def convert_y(dataY):
    y = []
    for i in range (len(dataY)):
        if(dataY[i] > 100):
            y.append(0)
        else:
            y.append(1)
    return np.array(y)
```
